### 1. Basic Statistics

Use Python to read in your data file and show many rows and columns your data has.

```python
import pandas as pd

# Load the data file
file_path = '/content/matches.csv'
data = pd.read_csv(file_path)

# Display the number of rows and columns
num_rows, num_columns = data.shape
print(data.head())
print(f"The dataset contains {num_rows} rows and {num_columns} columns")
```

### 2. Data Diversity

Use Python to identify how many unique values each column in your data has.  Use Python to identify which column has the greatest number of distinct values (aka the most diversity)?  If your data has a unique identifier (e.g. a person or visit or record ID) then find the column with the next most number of unique values.

```python
# Calculate the number of unique values for each column
unique_values = data.nunique()
print("Unique values per column:")
print(unique_values)

# Identify the column with the greatest number of distinct values
most_unique_col = unique_values.idxmax()
print(f"The column with the most distinct values is '{most_unique_col}' with {unique_values[most_unique_col]} unique values.")

# Check if it's a unique identifier
if unique_values[most_unique_col] == num_rows:
    next_most_unique_col = unique_values.drop(index=most_unique_col).idxmax()
    print(f"The column with the next most unique values is '{next_most_unique_col}' with {unique_values[next_most_unique_col]} unique values.")
```

### 3. Data Redundancy

Use Python to identify any columns in your data where the value for every row is the same.  For example, if you had a gender column and all the rows said 'M', then you would want to write the code that helped you determine that.

```python
# Identify columns with the same value for all rows
redundant_columns = unique_values[unique_values == 1].index.tolist()

if redundant_columns:
    print(f"The following columns have the same value for all rows: {redundant_columns}")
else:
    print("No columns have the same value for all rows.")
```

### 4. Range of Values

Your data almost certainly has some numeric columns that can be summed or averaged.  Create a histogram that shows the distribution of values for this column.

```python
import matplotlib.pyplot as plt

# Identify numeric columns
numeric_columns = data.select_dtypes(include='number').columns

if not numeric_columns.empty:
    print("Numeric columns identified for histogram generation:")
    print(numeric_columns)

    # Generate histograms for each numeric column
    for col in numeric_columns:
        plt.figure()
        data[col].hist(bins=20, edgecolor='black')
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.show()
else:
    print("No numeric columns found for histogram generation.")
```
